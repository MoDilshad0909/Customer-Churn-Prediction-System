## Data Cleaning

```python
# Convert TotalCharges to numeric, coercing errors to NaN
df['Total Charges'] = pd.to_numeric(df['Total Charges'].replace(' ', np.nan), errors='coerce')
```

```python
# Handle null values (drop rows with missing TotalCharges)
df.dropna(subset=['Total Charges'], inplace=True)
```

```python
# Remove duplicate rows if any
df.drop_duplicates(inplace=True)
```

```python
# Remove CustomerID column
if 'CustomerID' in df.columns:
    df.drop('CustomerID', axis=1, inplace=True)
```

### Churn Distribution

```python

# Calculate churn distribution
churn_counts = df['Churn Label'].value_counts().reset_index()
churn_counts.columns = ['Churn Label', 'Count']

# Create pie chart using Plotly
fig = px.pie(churn_counts, values='Count', names='Churn Label', 
             title='Churn Distribution', 
             color='Churn Label',
             color_discrete_map={'Yes':'#ef553b', 'No':'#636efa'})

# Update layout for professional look
fig.update_layout(title_x=0.5, template='plotly_white')
fig.show()

# Save as high-resolution PNG
fig.write_image('../images/churn_distribution.png', scale=2)
```

**Business Insight:**
The dataset shows a significant imbalance, with a notable portion of customers churning. Retaining these customers should be a primary objective for the business.

### Categorical Feature Countplots

```python

# List of selected categorical features
cat_features = ['Online Security', 'Tech Support', 'Streaming TV', 'Streaming Movies']

fig = make_subplots(rows=2, cols=2, subplot_titles=cat_features)

for i, feature in enumerate(cat_features):
    row = (i // 2) + 1
    col = (i % 2) + 1
    
    # Calculate counts
    counts = df[feature].value_counts().reset_index()
    counts.columns = [feature, 'Count']
    
    fig.add_trace(go.Bar(x=counts[feature], y=counts['Count'], name=feature), row=row, col=col)

fig.update_layout(title_text='Categorical Feature Countplots', title_x=0.5, template='plotly_white', showlegend=False)
fig.show()

fig.write_image('../images/categorical_feature_countplots.png', scale=2)
```

**Business Insight:**
A significant number of customers do not have supplementary services like Online Security or Tech Support. Bundling these services could be an effective strategy to reduce churn.

### Churn by Tenure Groups

```python

# Create tenure groups
bins = [0, 12, 24, 36, 48, 60, 72]
labels = ['0-1 year', '1-2 years', '2-3 years', '3-4 years', '4-5 years', '5-6 years']
df['Tenure Group'] = pd.cut(df['Tenure Months'], bins=bins, labels=labels, right=False)

tenure_group_churn = df.groupby(['Tenure Group', 'Churn Label']).size().reset_index(name='Count')

fig = px.bar(tenure_group_churn, x='Tenure Group', y='Count', color='Churn Label', barmode='group',
             title='Churn by Tenure Groups',
             color_discrete_map={'Yes':'#ef553b', 'No':'#636efa'},
             labels={'Count': 'Number of Customers'})
             
fig.update_layout(title_x=0.5, template='plotly_white')
fig.show()

fig.write_image('../images/churn_by_tenure_groups.png', scale=2)
```

**Business Insight:**
The '0-1 year' tenure group experiences the highest churn volume. Early engagement and onboarding programs are critical to improving customer lifetime value.

### Churn by Monthly Charges

```python

# Create monthly charge groups
charge_bins = [0, 30, 60, 90, 120]
charge_labels = ['Low ($0-$30)', 'Medium ($30-$60)', 'High ($60-$90)', 'Premium ($90+)']
df['Monthly Charge Group'] = pd.cut(df['Monthly Charges'], bins=charge_bins, labels=charge_labels, right=False)

charge_group_churn = df.groupby(['Monthly Charge Group', 'Churn Label']).size().reset_index(name='Count')

fig = px.bar(charge_group_churn, x='Monthly Charge Group', y='Count', color='Churn Label', barmode='group',
             title='Churn by Monthly Charges',
             color_discrete_map={'Yes':'#ef553b', 'No':'#636efa'},
             labels={'Count': 'Number of Customers'})
             
fig.update_layout(title_x=0.5, template='plotly_white')
fig.show()

fig.write_image('../images/churn_by_monthly_charges.png', scale=2)
```

**Business Insight:**
Customers in the 'High ($60-$90)' and 'Premium ($90+)' brackets exhibit significant churn. A review of the value proposition for high-tier plans is recommended.

### Feature Importance Preview

```python

# Ensure Churn Value exists
if 'Churn Value' not in df.columns:
    df['Churn Value'] = df['Churn Label'].map({'Yes': 1, 'No': 0})

# Select numeric features for correlation
numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
if 'Churn Value' in numeric_features:
    corr_with_churn = df[numeric_features].corr()['Churn Value'].sort_values(ascending=False)
    # Drop Churn Value itself
    corr_with_churn = corr_with_churn.drop('Churn Value', errors='ignore')
    
    # Create DataFrame for plotting
    corr_df = corr_with_churn.reset_index()
    corr_df.columns = ['Feature', 'Correlation with Churn']
    
    fig = px.bar(corr_df, x='Correlation with Churn', y='Feature', orientation='h',
                 title='Feature Importance Preview (Linear Correlation)',
                 color='Correlation with Churn', color_continuous_scale='RdBu_r')
                 
    fig.update_layout(title_x=0.5, template='plotly_white', yaxis={'categoryorder':'total ascending'})
    fig.show()
    
    fig.write_image('../images/feature_importance_preview.png', scale=2)
else:
    print("Could not compute feature importance: Target numeric column missing.")
```

**Business Insight:**
Features such as Monthly Charges and CLTV display a positive correlation with churn, while Tenure shows a strong negative correlation. These insights will directly inform future predictive modeling.
